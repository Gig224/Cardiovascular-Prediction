```python
# Core libraries
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Preprocessing and ML
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import Pipeline
from sklearn.inspection import permutation_importance

# Handling imbalance
from imblearn.over_sampling import SMOTE

# Explainability
import shap
import lime
import lime.lime_tabular

# Warnings
import warnings
warnings.filterwarnings('ignore')

# Style
sns.set(style='whitegrid')
```

```python
# Load the dataset
df = pd.read_csv("cardio_train.csv", sep=";")  # The Sulianova dataset uses semicolons

# Show basic structure
print("Shape of dataset:", df.shape)
df.head()
# Show first 10 and last 10 rows
pd.concat([df.head(10), df.tail(10)])
```

```python
# Summary info
df.info()

# Check for missing values
df.isnull().sum()
```

```python
# Descriptive stats
df.describe()

# Distribution of target class
sns.countplot(x="cardio", data=df)
plt.title("Target Variable Distribution (0 = No CVD, 1 = CVD)")
plt.xlabel("Cardiovascular Disease")
plt.ylabel("Count")
plt.show()

# Percentage breakdown
df["cardio"].value_counts(normalize=True) * 100
```

```python
# Drop the 'id' column if present
if 'id' in df.columns:
    df.drop('id', axis=1, inplace=True)

# Separate features (X) and target (y)
X = df.drop('cardio', axis=1)
y = df['cardio']

# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

print("Train-Test Split Complete")
```

```python
# StandardScaler normalization
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

from imblearn.over_sampling import SMOTE

# Apply SMOTE to the training set only
smote = SMOTE(random_state=42)
X_train_bal, y_train_bal = smote.fit_resample(X_train_scaled, y_train)


# Check new class distribution
sns.countplot(x=y_train_bal)
plt.title("Balanced Target Distribution After SMOTE")
plt.xlabel("Cardiovascular Disease")
plt.ylabel("Count")
plt.show()
```

```python
# Logistic Regression
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_train_bal, y_train_bal)

# Predictions
y_pred_log = log_reg.predict(X_test_scaled)

# Evaluation
print("📊 Logistic Regression Report")
print(classification_report(y_test, y_pred_log))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Train Random Forest on SMOTE-balanced data
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_bal, y_train_bal)

# Predict on original test set
y_pred_rf = rf.predict(X_test_scaled)

# Evaluate
print("📊 Random Forest Report")
print(classification_report(y_test, y_pred_rf))
```

```python
# Neural Network
mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=500, random_state=42)
mlp.fit(X_train_bal, y_train_bal)

# Predictions
y_pred_mlp = mlp.predict(X_test_scaled)

# Evaluation
print("📊 Neural Network Report")
print(classification_report(y_test, y_pred_mlp))
```

```python
# compare model accuracy scores
acc_log = accuracy_score(y_test, y_pred_log)
acc_rf = accuracy_score(y_test, y_pred_rf)
acc_mlp = accuracy_score(y_test, y_pred_mlp)

# Plot
plt.figure(figsize=(6, 4))
plt.bar(['Logistic Regression', 'Random Forest', 'Neural Network'], 
        [acc_log, acc_rf, acc_mlp], color=['skyblue', 'limegreen', 'salmon'])
plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy")
plt.ylim(0.7, 1)
plt.show()
```

```python
# define evaulation metrics
def evaluate_model(y_true, y_pred, model_name):
    print(f"\n📌 {model_name}")
    print("Accuracy:", round(accuracy_score(y_true, y_pred), 4))
    print("Precision:", round(precision_score(y_true, y_pred), 4))
    print("Recall:", round(recall_score(y_true, y_pred), 4))
    print("F1 Score:", round(f1_score(y_true, y_pred), 4))
    print("ROC AUC:", round(roc_auc_score(y_true, y_pred), 4))

# Logistic Regression
evaluate_model(y_test, y_pred_log, "Logistic Regression")

# Random Forest
evaluate_model(y_test, y_pred_rf, "Random Forest")

# Neural Network
evaluate_model(y_test, y_pred_mlp, "Neural Network")
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

# Plot confusion matrix for Random Forest (best model usually)
ConfusionMatrixDisplay.from_estimator(rf, X_test_scaled, y_test, cmap='Blues')
plt.title("Confusion Matrix - Random Forest")
plt.show()
```
