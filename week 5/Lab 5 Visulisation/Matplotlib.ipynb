```python
import numpy as np
import matplotlib.pyplot as plt

# Randomly generate data
np.random.seed(42)  # For reproducibility
height = np.random.normal(165, 10, 50)
weight = np.random.normal(70, 15, 50)

# Assign categories (e.g., Male and Female)
categories = np.random.choice(['Male', 'Female'], size=50)

# Assign colors based on categories
colors = ['blue' if category == 'Male' else 'red' for category in categories]

# Scatter plot with color coding
plt.scatter(height, weight, c=colors, alpha=0.6, edgecolor='black')
plt.title("Height vs. Weight (Categorized by Gender)")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")

# Add legend
plt.scatter([], [], color='blue', label='Male', alpha=0.6, edgecolor='black')
plt.scatter([], [], color='red', label='Female', alpha=0.6, edgecolor='black')
plt.legend()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example of multiple subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Line plot
axes[0].plot([1, 2, 3], [1, 4, 9], label='y = x^2', color='blue')
axes[0].set_title("Line Plot")
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")
axes[0].legend()

# Scatter plot
axes[1].scatter([1, 2, 3], [1, 4, 9], color='red', label='Points')
axes[1].set_title("Scatter Plot")
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [150, 200, 250, 300, 350]
profits = [50, 80, 100, 150, 200]

fig, ax1 = plt.subplots()

# Bar plot for sales
ax1.bar(months, sales, color='blue', label='Sales')
ax1.set_ylabel("Sales ($)")
ax1.set_xlabel("Months")
ax1.legend(loc='upper left')

# Line plot for profits
ax2 = ax1.twinx()
ax2.plot(months, profits, color='green', label='Profit', linestyle='--', marker='o')
ax2.set_ylabel("Profit ($)")
ax2.legend(loc='upper right')

plt.title("Sales and Profit Over Months")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset from the CSV file
file_path = r"C:\Users\X.Lu\COM7003 AI Labs\Lab 5 Visulisation\students_scores.csv"
df = pd.read_csv(file_path)

# Step 2: Line Plot
plt.plot(df['Hours_Studied'], df['Score'], marker='o', linestyle='--', color='blue')
plt.title("Hours Studied vs. Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Scores")
plt.grid(True)
plt.show()

# Step 3: Scatter Plot
plt.scatter(df['Hours_Studied'], df['Score'], color='green', alpha=0.7, edgecolor='black')
plt.title("Hours Studied vs. Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Scores")
plt.show()

# Step 4: Bar Chart
avg_scores = df.groupby("Subject")["Score"].mean()
plt.bar(avg_scores.index, avg_scores, color=['orange', 'purple', 'cyan', 'red'], edgecolor='black')
plt.title("Average Scores by Subject")
plt.xlabel("Subject")
plt.ylabel("Average Score")
plt.xticks(rotation=45)
plt.show()

# Step 5: Histogram
plt.hist(df['Score'], bins=10, color='skyblue', edgecolor='black')
plt.title("Distribution of Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.show()

# Step 6: Box Plot
subjects = df["Subject"].unique()
box_data = [df[df["Subject"] == subject]["Score"] for subject in subjects]

plt.boxplot(box_data, labels=subjects, patch_artist=True,
            boxprops=dict(facecolor="lightblue", color="black"),
            medianprops=dict(color="red", linewidth=2),
            whiskerprops=dict(color="black"))
plt.title("Scores by Subject")
plt.xlabel("Subject")
plt.ylabel("Score")
plt.show()

# Step 7: Correlation Heatmap (using Matplotlib only)
correlation = df[["Score", "Hours_Studied"]].corr()
fig, ax = plt.subplots()
im = ax.imshow(correlation, cmap="coolwarm", interpolation="nearest")

# Add labels
ax.set_xticks([0, 1])
ax.set_yticks([0, 1])
ax.set_xticklabels(["Score", "Hours_Studied"])
ax.set_yticklabels(["Score", "Hours_Studied"])

# Add values to heatmap cells
for i in range(correlation.shape[0]):
    for j in range(correlation.shape[1]):
        text = ax.text(j, i, f"{correlation.iloc[i, j]:.2f}",
                       ha="center", va="center", color="black")

plt.title("Correlation Heatmap")
plt.colorbar(im)
plt.show()
```
