```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load Dataset
df = pd.DataFrame({
    'Hours_Studied': [1, 2, 3, 4, 5],
    'Score': [40, 50, 60, 70, 80]
})

# Compute Correlation
correlation = df.corr()

# Create Heatmap
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
