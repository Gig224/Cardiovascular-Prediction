```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Step 1: Generate Synthetic Data
np.random.seed(42)  # For reproducibility
X = 2 * np.random.rand(100, 1)  # 100 random values between 0 and 2
y = 4 + 3 * X + np.random.randn(100, 1)  # y = 4 + 3X + noise

# Plot the data
plt.scatter(X, y, color="blue", label="Data points")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Generated Data for Linear Regression")
plt.legend()
plt.show()

# Step 2: Train Linear Regression Model using Scikit-Learn (Analytical Solution)
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_pred_sklearn = lin_reg.predict(X)

# Compute error metrics for Scikit-Learn model
mse_sklearn = mean_squared_error(y, y_pred_sklearn)
mae_sklearn = mean_absolute_error(y, y_pred_sklearn)
r2_sklearn = r2_score(y, y_pred_sklearn)

# Print results for Scikit-Learn Model
print("\nLinear Regression using Scikit-Learn:")
print(f"Optimized slope (m): {lin_reg.coef_[0][0]}")
print(f"Optimized intercept (b): {lin_reg.intercept_[0]}")
print(f"Mean Squared Error (MSE): {mse_sklearn}")
print(f"Mean Absolute Error (MAE): {mae_sklearn}")
print(f"R-squared Score: {r2_sklearn}")

# Step 3: Plot Regression Lines for Models

plt.figure(figsize=(8, 5))
plt.scatter(X, y, color="blue", label="Data points")
plt.plot(X, y_pred_sklearn, color="green", label="Scikit-Learn Regression Line")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression using Scikit-Learn")
plt.legend()
plt.show()

```
