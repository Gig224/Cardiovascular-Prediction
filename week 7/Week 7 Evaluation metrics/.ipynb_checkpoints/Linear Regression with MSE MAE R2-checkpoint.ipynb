```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Step 1: Generate Synthetic Data
np.random.seed(42)  # For reproducibility
X = 2 * np.random.rand(100, 1)  # 100 random values between 0 and 2
y = 4 + 3 * X + np.random.randn(100, 1)  # y = 4 + 3X + noise

# Plot the data
plt.scatter(X, y, color="blue", label="Data points")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Generated Data for Linear Regression")
plt.legend()
plt.show()

# Step 2: Train Linear Regression Model using Scikit-Learn
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_pred_mse = lin_reg.predict(X)

# Step 3: Compute MSE & MAE Errors
mse = mean_squared_error(y, y_pred_mse)
mae = mean_absolute_error(y, y_pred_mse)

# Print Results
print(f"Linear Regression using MSE:")
print(f"Optimized slope (m): {lin_reg.coef_[0][0]}")
print(f"Optimized intercept (b): {lin_reg.intercept_[0]}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Mean Absolute Error (MAE): {mae}")

# Step 4: Train a Model Optimized for MAE using SGDRegressor
from sklearn.linear_model import SGDRegressor

sgd_reg_mae = SGDRegressor(loss="epsilon_insensitive", penalty=None, learning_rate="constant", eta0=0.1, max_iter=1000, tol=1e-3)
sgd_reg_mae.fit(X, y.ravel())  # Train using MAE-based optimization
y_pred_mae = sgd_reg_mae.predict(X)

# Print Results for MAE Model
print("\nLinear Regression using MAE:")
print(f"Optimized slope (m): {sgd_reg_mae.coef_[0]}")
print(f"Optimized intercept (b): {sgd_reg_mae.intercept_[0]}")

# Step 5: Plot Separate Regression Lines for MSE & MAE

# MSE Regression Line
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color="blue", label="Data points")
plt.plot(X, y_pred_mse, color="red", label="MSE Regression Line")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression using MSE")
plt.legend()
plt.show()

# MAE Regression Line
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color="blue", label="Data points")
plt.plot(X, y_pred_mae, color="green", label="MAE Regression Line")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression using MAE")
plt.legend()
plt.show()
```
